```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '1719219914-Analysis of Super Store - DA (1).csv'
superstore_data = pd.read_csv(file_path)

# Set the style for the plots
sns.set(style="whitegrid")
```

### Overall Performance


```python
# Overall Performance
total_sales = superstore_data['Sales'].sum()
total_profit = superstore_data['Profit'].sum()

print(f"Total Sales: {total_sales}")
print(f"Total Profit: {total_profit}")

# Pie chart for sales and profit
labels = 'Sales', 'Profit'
sizes = [total_sales, total_profit]
colors = ['skyblue', 'lightgreen']
explode = (0.1, 0)  # explode the first slice

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%',
        shadow=True, startangle=140)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Total Sales vs Total Profit')
plt.show()
```

### Regional Performance


```python
# Regional Performance
region_sales = superstore_data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_profit = superstore_data.groupby('Region')['Profit'].sum().sort_values(ascending=False)

print(region_sales)
print(region_profit)

# Bar charts for sales and profit by region
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x=region_sales.index, y=region_sales.values, ax=ax[0], palette='Blues_d')
ax[0].set_title('Sales by Region')
ax[0].set_xlabel('Region')
ax[0].set_ylabel('Sales')

sns.barplot(x=region_profit.index, y=region_profit.values, ax=ax[1], palette='Greens_d')
ax[1].set_title('Profit by Region')
ax[1].set_xlabel('Region')
ax[1].set_ylabel('Profit')

plt.tight_layout()
plt.show()

```

### Category and Sub-Category Performance


```python
# Category and Sub-Category Performance
category_sales = superstore_data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_profit = superstore_data.groupby('Category')['Profit'].sum().sort_values(ascending=False)

subcategory_sales = superstore_data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
subcategory_profit = superstore_data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

print(category_sales)
print(category_profit)
print(subcategory_sales)
print(subcategory_profit)

# Bar charts for sales and profit by category and sub-category
fig, ax = plt.subplots(2, 2, figsize=(16, 12))

sns.barplot(x=category_sales.index, y=category_sales.values, ax=ax[0, 0], palette='Blues_d')
ax[0, 0].set_title('Sales by Category')
ax[0, 0].set_xlabel('Category')
ax[0, 0].set_ylabel('Sales')

sns.barplot(x=category_profit.index, y=category_profit.values, ax=ax[0, 1], palette='Greens_d')
ax[0, 1].set_title('Profit by Category')
ax[0, 1].set_xlabel('Category')
ax[0, 1].set_ylabel('Profit')

sns.barplot(x=subcategory_sales.index, y=subcategory_sales.values, ax=ax[1, 0], palette='Blues_d')
ax[1, 0].set_title('Sales by Sub-Category')
ax[1, 0].set_xlabel('Sub-Category')
ax[1, 0].set_ylabel('Sales')
ax[1, 0].tick_params(axis='x', rotation=90)

sns.barplot(x=subcategory_profit.index, y=subcategory_profit.values, ax=ax[1, 1], palette='Greens_d')
ax[1, 1].set_title('Profit by Sub-Category')
ax[1, 1].set_xlabel('Sub-Category')
ax[1, 1].set_ylabel('Profit')
ax[1, 1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
